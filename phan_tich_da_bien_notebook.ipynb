nhập dữ liệu

```python
df_diabetes = pd.read_csv('data/pima-indians-diabetes.csv', header=None, names=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'])
```

 #Kiểm tra thông tin cơ bản để xem có cột nào bị thiếu dữ liệu không

```python
print(df_diabetes.info())
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Vẽ histogram cho tất cả các biến

```python
df_diabetes.hist(figsize=(12, 10), bins=20)
plt.suptitle("Phân phối của từng đặc trưng")
plt.show()
```

# 2. Vẽ Boxplot để xem outliers (Ví dụ cho cột BMI)

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_diabetes['BMI'])
plt.title("Kiểm tra outliers trong cột BMI")
plt.show()
```

```python
# So sánh Glucose giữa người bị và không bị tiểu đường
sns.boxplot(x='Outcome', y='Glucose', data=df_diabetes)
plt.title("So sánh mức Glucose giữa 2 nhóm")
plt.show()
```
