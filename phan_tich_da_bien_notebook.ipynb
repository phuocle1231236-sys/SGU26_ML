# Tại sao Terminal không hiển thị ảnh và cách khắc phục

## 1. Hiểu về việc hiển thị biểu đồ trong Jupyter Notebooks

Jupyter notebooks hiển thị biểu đồ trực tiếp trong cell output nhờ backend 'inline', nhưng khi chạy script Python trong terminal, `plt.show()` thường không hiển thị ảnh vì terminal không hỗ trợ giao diện đồ họa (GUI). Thay vào đó, bạn có thể thấy thông báo lỗi hoặc không có gì xảy ra.

## 2. Chạy Jupyter trong chế độ Terminal

Khi chạy notebook trong terminal (ví dụ: `jupyter notebook`), `plt.show()` có thể không hiển thị nếu không có GUI backend. Bạn cần cấu hình matplotlib để sử dụng backend phù hợp.

## 3. Cấu hình Matplotlib Backend

Để tránh lỗi khi chạy trong môi trường không GUI, đặt backend thành 'Agg' và lưu biểu đồ thành file thay vì hiển thị.

```python
import matplotlib
matplotlib.use('Agg')  # Backend không GUI
import matplotlib.pyplot as plt

# Ví dụ: Lưu biểu đồ thay vì show
plt.plot([1, 2, 3], [1, 4, 2])
plt.savefig('plot.png')  # Lưu thành file
print("Biểu đồ đã lưu thành plot.png")
```

## 4. Sử dụng Inline Backend cho VS Code

Trong VS Code với Jupyter extension, sử dụng `%matplotlib inline` để hiển thị biểu đồ trực tiếp trong cell output.

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Ví dụ: Biểu đồ sẽ hiển thị inline
plt.plot([1, 2, 3], [1, 4, 2])
plt.show()
```
